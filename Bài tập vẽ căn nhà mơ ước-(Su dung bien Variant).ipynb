```python
import turtle

#biến màu
bg_color='cyan'
house_front_color='blue'
house_side_color='yellow'
house_front_roof_color='orange'
house_side_roof_color='green'
house_door_color='grey'
house_window_color='brown'


#setup màn hình nền
t=turtle
t.Screen()
t.setup(800,680)
t.bgcolor(bg_color)

#setup bút vẽ
pen=turtle.Turtle()

#Toạ độ gốc căn nhà
x=-150
y=-250



#Biến số vẽ góc độ
goc_do_1 = 90
goc_do_2 =60



#vẽ căn nhà
chiều_dài_mặt_trước_nhà=140
chiều_cao_mặt_trước_nhà=170
pen.speed(10)
pen.begin_fill()
pen.color(house_front_color)
pen.penup()
pen.goto(x,y)
pen.pendown()
pen.backward(chiều_dài_mặt_trước_nhà)
pen.left(goc_do_1)
pen.forward(chiều_cao_mặt_trước_nhà)
pen.right(goc_do_1)
pen.forward(chiều_dài_mặt_trước_nhà)
pen.right(goc_do_1)
pen.forward(chiều_cao_mặt_trước_nhà)
pen.end_fill()

#vẽ nóc nhà trước→OK
pen.speed(10)
pen.begin_fill()
pen.color(house_front_roof_color)
pen.penup()
pen.goto(x,-80)
pen.pendown()
pen.right(goc_do_1)
for x in range(3):    
    pen.forward(chiều_dài_mặt_trước_nhà)
    pen.right(120)
pen.end_fill()


#vẽ nóc nhà bên hông

pen.speed(10)
pen.begin_fill()
pen.color(house_side_roof_color)

pen.right(goc_do_2)
pen.forward(chiều_dài_mặt_trước_nhà)
pen.right(goc_do_1)
pen.forward(chiều_cao_mặt_trước_nhà)
pen.right(goc_do_1)
pen.forward(chiều_dài_mặt_trước_nhà)
pen.right(goc_do_1)
pen.forward(chiều_cao_mặt_trước_nhà)

pen.end_fill()

#vẽ bên hông nhà
pen.speed(5)
pen.color(house_side_color)
pen.begin_fill()
pen.left(goc_do_2)
pen.forward(chiều_cao_mặt_trước_nhà)
pen.left(120)
pen.forward(chiều_cao_mặt_trước_nhà)
pen.left(goc_do_2)
pen.forward(chiều_cao_mặt_trước_nhà)
pen.end_fill()

#vẽ cửa căn nhà
pen.speed(10)
pen.up()
pen.goto(-250,-250)
pen.pendown()
pen.color(house_door_color)
pen.begin_fill()
pen.forward(100)
pen.right(goc_do_1)
pen.forward(goc_do_2)
pen.right(goc_do_1)
pen.forward(100)
pen.right(goc_do_1)
pen.forward(goc_do_2)
pen.end_fill()

#vẽ cửa sổ
pen.speed(10)
pen.up()
pen.goto(-90,-90)
pen.down()
pen.color(house_window_color)
pen.begin_fill()
pen.left(goc_do_1)
pen.forward(50)
pen.left(120)
pen.forward(50)
pen.left(goc_do_2)
pen.forward(50)
pen.right(-120)
pen.forward(50)
pen.end_fill()

#vẽ cây
toạ_độ_cây_x=250
toạ_độ_cây_y=-260

pen.up()
pen.right(30)
pen.goto(toạ_độ_cây_x,toạ_độ_cây_y)
pen.down()
pen.speed(10)
pen.color(house_window_color)
pen.begin_fill()
pen.forward(50)
pen.right(goc_do_1)
pen.forward(100)
pen.right(goc_do_1)
pen.forward(50)
pen.end_fill()

#vẽ lá cây
pen.up()
pen.goto(150,-160)
pen.down()
pen.color('green')
pen.begin_fill()
for x in range(3):
    pen.forward(150)
    pen.left(120)
pen.end_fill() 

pen.up()
pen.goto(165,-30)
pen.down()
pen.color('green')
pen.begin_fill()
for x in range(3):
    pen.forward(120)
    pen.left(120)
pen.end_fill()    

pen.up()
pen.goto(180,73)
pen.down()
pen.color('green')
pen.begin_fill()
for x in range(3):
    pen.forward(100)
    pen.left(120)
pen.end_fill()   

#vẽ ông mặt trời

pen.up()
pen.goto(100,170)
pen.down()
pen.color('yellow')
pen.begin_fill()
pen.circle(70)
pen.end_fill()

pen.up()
pen.goto(100,170)
pen.color('black')
pen.width(3)
pen.down()
pen.right(goc_do_1)
pen.forward(80)

pen.up()
pen.goto(27,240)
pen.color('black')
pen.width(3)
pen.down()
pen.right(goc_do_1)
pen.forward(80)

pen.up()
pen.goto(100,310)
pen.color('black')
pen.width(3)
pen.down()
pen.right(goc_do_1)
pen.forward(80)

pen.up()
pen.goto(171,240)
pen.color('black')
pen.width(3)
pen.down()
pen.right(goc_do_1)
pen.forward(80)


pen.up()
pen.goto(151,190)
pen.color('black')
pen.width(3)
pen.down()
pen.right(45)
pen.forward(40)

pen.up()
pen.goto(51,190)
pen.color('black')
pen.width(3)
pen.down()
pen.right(goc_do_1)
pen.forward(40)

pen.up()
pen.goto(51,290)
pen.color('black')
pen.width(3)
pen.down()
pen.right(goc_do_1)
pen.forward(40)

pen.up()
pen.goto(151,290)
pen.color('black')
pen.width(3)
pen.down()
pen.right(goc_do_1)
pen.forward(40)

turtle.done()







```
